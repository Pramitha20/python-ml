# Train Test Split

Train Test Split is a fundamental technique of data preprocessing where it involves splitting the datasets into model training data and model testing data.

```python
from sklearn import datasets
data1=datasets.load_breast_cancer()
data1
```

```python
import numpy as np
import pandas as pd
data2=pd.DataFrame(data1.data,columns=data1.feature_names)
data2
```

```python
x = data2
y = data1.target
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
print(y.shape,y_train.shape,y_test.shape)
```
